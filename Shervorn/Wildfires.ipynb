```python
import os
import pandas as pd
import plotly.express as px
```

```python
df = pd.read_csv(r".\Data\CA_wildfires.csv")
df.head(5)
```

```python
df.shape
```

```python
df.info()
```

```python
colToDrop = [
    'attr_IncidentName', 'attr_FireCauseGeneral', 'attr_InitialResponseDateTime', 'attr_PredominantFuelGroup'
]

cleanedDF = df.drop(colToDrop, axis=1).drop_duplicates()

cleanedDF['attr_FireDiscoveryDateTime'] = cleanedDF['attr_FireDiscoveryDateTime'].str[:10]
cleanedDF["attr_FireDiscoveryDateTime"] = pd.to_datetime(cleanedDF["attr_FireDiscoveryDateTime"])

cleanedDF['attr_FireOutDateTime'] = cleanedDF['attr_FireOutDateTime'].str[:10]
cleanedDF["attr_FireOutDateTime"] = pd.to_datetime(cleanedDF["attr_FireOutDateTime"])

cleanedDF.head(5)
```
