```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import folium
from folium.plugins import MarkerCluster
```

```python
aqi = pd.read_csv('data/cleaned_AQI.csv')
df = pd.read_csv('data/CA_wildfires.csv')
```

```python
df.head()
```

```python
df = df.dropna(subset=['attr_InitialLatitude','attr_InitialLongitude'])
```

```python
# df['attr_InitialLatitude'].isna().sum()
df['attr_InitialLatitude'].isna().sum()
```

```python
center_lat = 36.7783
center_long = -119.4179
map = folium.Map(location=[center_lat,center_long], zoom_start=6)
marker_cluster = MarkerCluster().add_to(map)

for index, row in df.iterrows():
    lat = row['attr_InitialLatitude']
    long = row['attr_InitialLongitude']
    marker = folium.Marker([lat,long])
    popup = f'County: {df.attr_POOCounty}'
    marker.add_to(marker_cluster)

map
```
